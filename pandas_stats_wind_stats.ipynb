```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('wind.csv')
df.head()
```

```python
#df['year'] = pd.DatetimeIndex(df['Yr_Mo_Dy']).year
#df['month'] = pd.DatetimeIndex(df['Yr_Mo_Dy']).month
#df['day'] = pd.DatetimeIndex(df['Yr_Mo_Dy']).day
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)

df.head()
```

```python
df.set_index('Yr_Mo_Dy', inplace=True)
df.head()
```

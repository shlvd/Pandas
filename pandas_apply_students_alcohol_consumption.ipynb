```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student-mat.csv')
df.head()
```

```python
df = df.loc[:, 'school':'guardian']
df.head()
```

```python
cap = lambda x: x.capitalize()
```

```python
df['Mjob'] = df['Mjob'].apply(cap)
df['Fjob'] = df['Fjob'].apply(cap)
```

```python
def majority(df):
    df['legal_drinker'] = df.age > 17
    return df

majority(df).head()
```

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
    
df['legal_drinker'] = df['age'].apply(majority)
df.head()
```

```python
def times10(x):
    if type(x) is int:
        return 10 * x
    return x

df.applymap(times10).head(5)
```

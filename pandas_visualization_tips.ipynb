```python
import pandas as pd

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns


# print the graphs in the notebook
% matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

```python
df = pd.read_csv('tips.csv')
del df['Unnamed: 0']
df.head()
```

```python
df.total_bill.hist(bins=20)
```

```python
ttlbill = sns.distplot(df.total_bill)
ttlbill.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill")
#sns.despine()
```

```python
plt.scatter(df.total_bill, df.tip)
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = df)
```

```python
sns.pairplot(df)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = df, jitter = True)
```

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = df, jitter = True)
```

```python
sns.boxplot(x='total_bill', y='day', data=df, orient='h')
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = df)
```

```python
sns.set(style = "ticks")
g = sns.FacetGrid(df, col = "time")
g.map(plt.hist, "tip")
```

```python
g = sns.FacetGrid(df, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7)
g.add_legend()
```
